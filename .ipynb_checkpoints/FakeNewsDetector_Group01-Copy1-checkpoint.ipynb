```python
#Below should be replaced with actual data from after Cleaning
cleaned_data_SAMPLE_FOR_CODING_PURPOSES = "<url> <url> what are the <number> strands of our dna and why is a war against our dna? trailer for awakening of <number> strands the full video is only available as a paid video on vimeo"
#Create tokens based on clean_data. cl = clean, da = data
clda_tokens = word_tokenize(cleaned_data_SAMPLE_FOR_CODING_PURPOSES)

#print((clda_tokens))
```

```python
#StopWordsCLeanDAta_tokens
swclda_tokens = [word for word in clda_tokens if not word in stopwords.words('english')]

```

```python
stemmer = porter.PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in swclda_tokens]
#create Stemmed StopWorded vocab
stsw_vocab = set(stemmed_tokens)
stsw_vocab.remove('<')
stsw_vocab.remove('>')
print(stsw_vocab)
```
